ONE HOT ENCODING(A technique that converts each unique word into a binary vector where only one position is 1 and all others are 0.)

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
doc=['people watch movie',
          'people watch cricket',
          'people like movie',
          'people like cricket'] 
```

```python
tokens=[sentence.lower().split() for sentence in doc]
print(tokens)
```

```python
words=[[word]for sentence in tokens for word in sentence]
print(words)
```

```python
encoder.fit(words)
```

```python
print("vocabulary:",encoder.categories_[0])
```

```python
#covert data into one hot encoded format
for sentence in tokens:
    encoded=encoder.transform([[word] for word in sentence]).toarray()
    print("orignalsentence:",sentence)
    print(encoded)
```

BOW-BAG OF WORDS(It check how many time word present in the doc if it present more then on time in the documemt then its frequency is high)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow=CountVectorizer()
```

```python
bow.fit(doc)
```

```python
print("vocabulary:",bow.get_feature_names_out())
```

```python
BOW_Encoded=bow.transform(doc).toarray()
print("BOW Encoded:",BOW_Encoded)
```

TFIDF-TERM-FREQUENCY-INVERSE-DOCUMENT-FREQUENCY(It check how much word is important in the document if it is more important then its importance is high) 
Formula:
termcount/ total word in document*log(total_Doc/no of doc containing that term)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
tfidf.fit(doc)
```

```python
print("vocabulary:",tfidf.get_feature_names_out())
```

```python
tfidf_encoded=tfidf.transform(doc).toarray()
print("TFIDF Encoded:",tfidf_encoded)
```
