HANDLING THE PUNCTUAIONS

```python
import string
import pandas as pd
import numpy as np
```

```python
string.punctuation
```

```python
def remove_punctuation(text):
    exclude=string.punctuation
    for char in exclude:
        text=text.replace(char, "")
    return text
```

```python
text = "@myname is vansh!!!!!"
remove_punctuation(text)
```

```python
text1="fYI this is not true"
text2="LAMO the class was so funny"
text3="ASAP we need to finish this project"
```

```python
chat_abb = {
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "LAMO": "Laugh At My Own",
}
```

```python
chat_abb["ASAP"]
```

```python
def replace_abbreviations(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_abb:
            new_text.append(chat_abb[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)   
```

```python
replace_abbreviations(text3)
```

STOP WORDS

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
text1='hi i am vansh khanna and i am a student of btech in computer science and engineering'
```

```python
def remove_stopwords(text):
    new_text=[]
    for w in text.split():
        if w.lower() in stopwords.words("english"):
            new_text.append("")
        else:  
            new_text.append(w.strip())
    return " ".join(new_text)
```

```python
remove_stopwords(text1)
```

#tokenization using nltk library

```python
#Sent_tokenize
text1="Hello how are you. I am Vansh. I love AI."
from nltk.tokenize import sent_tokenize
sent_token=sent_tokenize(text1)
print(sent_token)
#word_tokenize
from nltk.tokenize import word_tokenize
text2="Hello how are you"
word_token=word_tokenize(text2)
print(word_token)
```

STEMMING(Stemming is the process of removing prefixes or suffixes from a word to get its root form. The resulting word may not be a valid dictionary word.)

```python
from nltk.stem import PorterStemmer
def stemming(text):
    obj = PorterStemmer()
    for w in text.split():
        print(f"Stem of {w} is {obj.stem(w)}")
    return stemming
```

```python
text="running runner runs"
stemming(text)
```

LEMMATIZATION(Lemmatization is the process of converting a word to its dictionary (base) form using vocabulary and grammar rules. The result is always a meaningful word.)

```python
from nltk.stem import WordNetLemmatizer
def lemmatization(text):
    obj=WordNetLemmatizer()
    for w in text.split():
        print(f"Lemmatization of {w} is {obj.lemmatize(w)}")
    return lemmatization
```

```python
text="The cats are chasing the mice. The children are running and playing in the parks. They studied hard and went to school every day."
lemmatization(text)
```
