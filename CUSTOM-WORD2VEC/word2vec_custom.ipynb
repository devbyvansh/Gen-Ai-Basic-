```python
#import libraries
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import nltk
from nltk.tokenize import sent_tokenize
import os
```

```python
#for loop is used to read both text and file and then tokenize the sentence and creates its embeddings
folder_path="D:\\GENAI\\GEN AI BASICS\\CUSTOM-WORD2VEC"
story_model=[]
for file in os.listdir(folder_path):
    if file.endswith('.txt'):
        file_path=os.path.join(folder_path,file)
        with open(file_path, encoding='utf-8') as f:
            corpus=f.read()
            raw_sent=sent_tokenize(corpus)
            for i in raw_sent: 
                story_model.append(simple_preprocess(i))
```

```python
#create a custom word2vec model using gensim library
custom_model=gensim.models.Word2Vec(sentences=story_model,vector_size=100, window=10, min_count=1)
```

```python
#create a custom vocabulary to prepare the model for training
custom_model.build_vocab(story_model)
```

```python
#train the model to learn the embeddings of the words in the vocabulary
custom_model.train(story_model, total_examples=custom_model.corpus_count, epochs=5)
```

```python
#use to check the embeddings of the word "king" in the custom model
custom_model.wv['king']
```

```python
#use to check the similarity between the words "monkey" and "rabbit" in the custom model
print(custom_model.wv.similarity("monkey", "rabbit"))
```

```python
#use to check the most similar words to "monkey" in the custom model
custom_model.wv.most_similar("monkey")
```
