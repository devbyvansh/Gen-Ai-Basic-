```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
google_model = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
```

```python
query="How can I improve the battery life of my smartphone?"
```

```python
query_embedding=google_model.embed_query(query)
```

```python
documents = [
    "Smartphone battery life can be improved by reducing screen brightness, enabling battery saver mode, and closing background apps.",

    "Modern mobile devices consume more power when GPS, Bluetooth, and Wi-Fi are continuously enabled.",

    "Charging your phone between 20% and 80% helps maintain long-term battery health.",

    "Artificial Intelligence is transforming healthcare through medical image analysis and disease prediction.",

    "The Indian Premier League is one of the most popular T20 cricket tournaments in the world.",

    "Mobile phones with OLED displays usually consume less power when using dark mode.",

    "Laptop battery optimization includes lowering screen brightness and disabling unnecessary startup programs.",

    "Regular software updates often include battery performance improvements and bug fixes."
]
```

```python
documents_embeddings = google_model.embed_documents(documents)
```

SEMANTIC-SEARCH

```python
import numpy as np
def dot_product(a,b):
    return np.dot(a,b)
def cosine_similarity(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
def euclidean_distance(a,b):
    a=np.array(a)
    b=np.array(b)
    return np.linalg.norm(a-b)
results=[]
for doc,doc_emb in zip (documents,documents_embeddings):
    results.append({
        "DOCUMENT":doc,
        "QUERY":query,
        "COSINE_SIMILARITY":cosine_similarity(query_embedding,doc_emb),
        "DOTPRODUCT":dot_product(query_embedding,doc_emb),
        "EUCLIDEAN_DISTANCE":euclidean_distance(query_embedding,doc_emb)
    })
```

KEYWORD-SEARCH

```python
from collections import Counter

def keyword_search(query, documents):
    values = []
    querysplit = query.lower().split()

    for doc in documents:
        docsplit = doc.lower().split()
        doc_counter = Counter(docsplit)

        score = sum(doc_counter[word] for word in querysplit)

        values.append({
            "DOCUMENT": doc,
            "QUERY": query,
            "KEYWORD_SCORE": score
        })

    # Sort by score (highest first)
    values = sorted(values, key=lambda x: x["KEYWORD_SCORE"], reverse=True)

    return values
```

```python
keyword_search(query, documents)
```
